```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/sales_clean.csv")
df['Dates'] = pd.to_datetime(df['Dates'])
df['Mois'] = df['Dates'].dt.to_period('M')
```

```python
sales_by_month = df.groupby('Mois')['Ventes'].sum()

plt.figure()
sales_by_month.plot()
plt.title("Evolution mensuelle du chiffre d'affaires")
plt.xlabel("Mois")
plt.ylabel("Chiffre d'Affaires")
plt.savefig("../visuals/Evolution mensuelle du chiffre d'affaires.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Ce que montre le graphique
Ce graphique linaire illustre l'volution mensuelle du chiffre d'affaires (CA) sur une priode de 24 mois, de janvier 2023  dcembre 2024. L'axe vertical montre le CA en euros (de 85 000  110 000+) et l'axe horizontal reprsente les mois successifs.
